### DataFrame 组成

```python
ndarray_data = np.array([
    ["甲", 25],
    ["乙", 30],
])

df = pandas.DataFrame(ndarray_data, columns=["name", "age"])
print(df)
```

```python
print(df.head(1))  # Display the first row of the DataFrame
print(df.tail(1))  # Display the last row of the DataFrame
```
